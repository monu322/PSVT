```
#print(df.columns)
psvt_df = df[:86]
non_psvt_df = df[86:]

x = psvt_df['ID'].values
y = psvt_df['TAS-20'].values
non_y = non_psvt_df['TAS-20'].values
#print(x)
#print(y)

plt.scatter(x, y, color='red', label='PSVT')
plt.scatter(x, non_y, color='green', label='non-PSVT')
plt.xlabel('ID')
# Set the y axis label of the current axis.
plt.ylabel('TAS-20')
# Set a title 
plt.title('TAS-20')
plt.legend()
plt.show()


```
